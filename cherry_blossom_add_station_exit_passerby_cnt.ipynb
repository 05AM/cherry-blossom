## 시간별 지하철 역 입구 통행자 정보 가져오기

```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv("SKT_API_KEY")
```

```python
url = "https://apis.openapi.sk.com/puzzle/subway/exit/raw/hourly/stations/133?gender=all&ageGrp=all&date=latest"

headers = {
    "accept": "application/json",
    "Content-Type": "application/json",
    "appkey": api_key
}

response = requests.get(url, headers=headers)

print(response.text)
```
