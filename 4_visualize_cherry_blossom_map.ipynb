```python
import pandas as pd

loc_data_0401 = pd.read_csv('source/cherryblossom_final_20230401.csv',
                     sep=',', engine='python')

loc_data_0402 = pd.read_csv('source/cherryblossom_final_20230402.csv',
                     sep=',', engine='python')

loc_data_0403 = pd.read_csv('source/cherryblossom_final_20230403.csv',
                     sep=',', engine='python')

loc_data_0404 = pd.read_csv('source/cherryblossom_final_20230404.csv',
                     sep=',', engine='python')

loc_data_0405 = pd.read_csv('source/cherryblossom_final_20230405.csv',
                     sep=',', engine='python')

loc_data_0406 = pd.read_csv('source/cherryblossom_final_20230406.csv',
                     sep=',', engine='python')

loc_data_0407 = pd.read_csv('source/cherryblossom_final_20230407.csv',
                     sep=',', engine='python')
```

## 지도로 시각화

```python
# 평일
data_average_weekday = loc_data_0401

data_average_weekday = data_average_weekday.drop('user_cnt_day', axis=1)
data_average_weekday = data_average_weekday.drop('user_cnt_night', axis=1)

# 주말
data_average_weekend = loc_data_0403

data_average_weekend = data_average_weekend.drop('user_cnt_day', axis=1)
data_average_weekend = data_average_weekend.drop('user_cnt_night', axis=1)
```

```python
data_average_weekday['average_user_cnt_day'] = (loc_data_0401['user_cnt_day'] + loc_data_0402['user_cnt_day'] + loc_data_0405['user_cnt_day'] + 
loc_data_0406['user_cnt_day'] + loc_data_0407['user_cnt_day']) / 5

data_average_weekday['average_user_cnt_night'] = (loc_data_0401['user_cnt_night'] + loc_data_0402['user_cnt_night'] + loc_data_0405['user_cnt_night'] + 
loc_data_0406['user_cnt_night'] + loc_data_0407['user_cnt_night']) / 5

data_average_weekday
```

```python
data_average_weekend['average_user_cnt_day'] = (loc_data_0403['user_cnt_day'] + loc_data_0404['user_cnt_day']) / 2

data_average_weekend['average_user_cnt_night'] = (loc_data_0403['user_cnt_night'] + loc_data_0404['user_cnt_night']) / 2

data_average_weekend
```

## 막대 그래프로 시각화

### 평일 밤

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# encoding
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

data_average_weekday = data_average_weekday.sort_values(by='average_user_cnt_night', ascending=False)

x = data_average_weekday['loc_name']
y = data_average_weekday['average_user_cnt_night']

plt.figure(figsize=(10, 8))

plt.bar(x, y)

# 그래프 제목과 축 레이블 설정
plt.title('평일 밤(17-23)')
plt.xlabel('location')
plt.ylabel('traffic')
plt.xticks(range(len(x)), x, rotation='vertical')

plt.savefig('result/day_exit_traffic_weekday_day.png')

plt.show()
```

### 주말 밤

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# encoding
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

data_average_weekend = data_average_weekend.sort_values(by='average_user_cnt_night', ascending=False)

x = data_average_weekend['loc_name']
y = data_average_weekend['average_user_cnt_night']

plt.figure(figsize=(10, 8))

plt.bar(x, y)

# 그래프 제목과 축 레이블 설정
plt.title('주말 밤(17-23)')
plt.xlabel('location')
plt.ylabel('traffic')
plt.xticks(range(len(x)), x, rotation='vertical')

plt.savefig('result/day_exit_traffic_weekend_night.png')

plt.show()
```

## 평일 낮

```python
import folium

map = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=12, 
#     tiles='cartodbpositron'
)

for index, row in data_average_weekday.iterrows():
    lat = row['lat']
    long = row['long']
    folium.CircleMarker([lat, long], 
                       radius = row['average_user_cnt_day']/400 + 10,
                       popup = folium.Popup("{}<br><br>{}".format(row['loc_name'], row['feature']), max_width=300, min_width=150),
                       color = "red",
                       fill = True).add_to(map)
    
map.save('result/map_cherry_blossom_weekday_day.html')

map
```
